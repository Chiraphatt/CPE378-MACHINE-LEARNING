```python
import nltk
nltk.download('punkt')
f = open('harry_7books.txt', encoding='utf8')
raw = f.read()
raw[:1000]
```

# 3. Tokenization

```python
tokenizer = nltk.RegexpTokenizer(r"\w+")
tokens= tokenizer.tokenize(raw)
print(tokens[3000:3100])
```

# 2. Case Folding

```python
tokens = [w.lower() for w in tokens]
```

```python
print(tokens[3000:3100])
```

# 4. Text normalization

```python
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
#print(stop_words)
```

```python
tokens = [w for w in tokens if not w in stop_words]
print(tokens[3000:3100])
```

```python
tokens = [w.replace('\xad','') for w in tokens]
tokens
```

```python
#lemmatize ทำให้คำที่มีความหมายคล้ายกันกลายเป็นคำเดียวกัน
nltk.download('wordnet')
wnl = nltk.WordNetLemmatizer()
tokens  = [wnl.lemmatize(w) for w in tokens]
```

```python
#Counter เอามานับจำนวนคำที่ซ้ำกันในเอกสารนี้
from collections import Counter
counter = Counter(tokens)
counter
```

```python
import pandas as pd
freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df.index.name = 'token'
freq_df.head(15)
```

```python
freqtok_df = freq_df.sort_values('freq', ascending=False)
freqtok_df.head(15)
```

```python
ax = freqtok_df.head(10).plot(kind='barh', width=0.8, figsize=(8,4))
ax.invert_yaxis()
ax.set(xlabel='Frequency', ylabel='Token', title='Top 10 Words in harry-potter dataset ')
```
